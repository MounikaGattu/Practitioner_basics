```python
#As of now lets delete the data instead of processing as it takes more time, usually we replace with mean values or so
#delete columns which has text (timesake)
data_final = data.drop(["Age", "Cabin", "Embarked", "Sex", "Name", "Ticket"], axis=1)
#axis=0, rows and axis=1, columns
```

```python
#size of data_final
data_final.shape
```

```python
#drop the dependent varaible from the set x, which contains independent varaibels
x= data_final.drop(["Survived"], axis=1)
```

```python
#define the dependnet varaible
y=data_final[["Survived"]]
```

```python
#train and test split
from sklearn.model_selection import train_test_split
```

```python
#model training and model testing (predefined syntax)
X_train, X_test, y_train, y_test = train_test_split(
x, y, test_size=0.10)
```

```python
#import different model pacakgeswhich you need to select and test
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
#from sklearn.svm import SVC
```

```python
#instance of model
lr = LogisticRegression()
knn = KNeighborsClassifier()
nb = GaussianNB()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
#fit
lr.fit(X_train, y_train)
knn.fit(X_train, y_train)
nb.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
#predict
p1=lr.predict(X_test)
p2=knn.predict(X_test)
p3=nb.predict(X_test)
p4=dt.predict(X_test)
p5=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#use print statement to print all the responses not just the last response
print(accuracy_score(p1,y_test))
print(accuracy_score(p2,y_test))
print(accuracy_score(p3,y_test))
print(accuracy_score(p4,y_test))
print(accuracy_score(p5,y_test))
```

```python
print(f1_score(p1,y_test))
print(f1_score(p2,y_test))
print(f1_score(p3,y_test))
print(f1_score(p4,y_test))
print(f1_score(p5,y_test))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
#display confusion matricx
confusion_matrix(p4,y_test)
```

```python
#import roc curve related packages
from sklearn.metrics import roc_auc_score, roc_curve
from matplotlib import pyplot
```

```python
probs = dt.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(y_test, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y_test, probs)
# plot no skill
pyplot.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
pyplot.plot(fpr, tpr, marker='.')
# show the plot
pyplot.show()
```
